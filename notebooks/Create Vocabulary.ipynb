```python
from pandas import read_pickle

df = read_pickle('../data/cuentalo_json_con_pais_ciudad_limpios.pkl')
df.sample(5)

df.describe()

df = df[df['tweet_type'] == 'original']
```

```python
import string
import unidecode import nltk
from nltk.stem import PorterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords

dict_of_words = dict()
ps = PorterStemmer()
sw=set(stopwords.words('spanish'))
count = 0

for text in df.full_text.values:
    text = str(text)
    for i in string.punctuation + '?' + "''" + '`' + '1234567890':
        text = text.replace(i, '')
        
    text = unidecode.unidecode(text)
    text = text.lower()    
    word_tk = word_tokenize(text)
    for word in word_tk:
        if word not in sw:
            
            if ps.stem(word) in dict_of_words.keys():
                dict_of_words[ps.stem(word)] = dict_of_words[ps.stem(word)] + 1

            else:
                dict_of_words[ps.stem(word)] = 1
    count += 1
    print('Tweet {} out of {}'.format(count, df.shape[0]))
print('There are {} words in this dictionary.'.format(len(dict_of_words)))
```

```python
dict_of_words['ao'] = dict_of_words.pop('ano')
dict_of_words['hombre'] = dict_of_words.pop('hombr')
dict_of_words['que'] = dict_of_words.pop('q')
dict_of_words['puedo'] = dict_of_words.pop('pued')
dict_of_words['despus'] = dict_of_words.pop('despu')
dict_of_words['noche'] = dict_of_words.pop('noch')
dict_of_words['madre'] = dict_of_words.pop('madr')
dict_of_words['veces'] = dict_of_words.pop('vece')
dict_of_words['nadie'] = dict_of_words.pop('nadi')
dict_of_words['siempre'] = dict_of_words.pop('siempr')
dict_of_words['padre'] = dict_of_words.pop('padr')
dict_of_words['parte'] = dict_of_words.pop('part')
dict_of_words['fuerte'] = dict_of_words.pop('fuert')
dict_of_words['coche'] = dict_of_words.pop('coch')
dict_of_words['triste'] = dict_of_words.pop('trist')
dict_of_words.pop('?')
dict_of_words.pop('``')
dict_of_words.pop("''")
```
